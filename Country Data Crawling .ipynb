```python
def CheckTime(func):
    import time

    def newFunc(*args, **kwargs):
        start = time.time()
        func(*args, **kwargs)
        end = time.time()
        
        print("\n실행시간 : ",end - start)

    return newFunc
```

```python
class WrongStudentRecord(Exception):
    def __init__(self, msg):
        self._message = msg
```

```python
@CheckTime
def get_population():
    population_crawling.age_structure()
    population_crawling.median_age()

@CheckTime
def get_Unemp():  
    try: 
        economy_crawling.Unemployment()
    except AttributeError:
        pass
        
    except IndexError:
        raise WrongStudentRecord("CountryNameCode.csv 파일의 나라 갯수를 다시 확인해주세요")
    
    except:
        pass 


@CheckTime
def get_GDP():
    try:
        economy_crawling.GDP_composition()
    except:
        raise WrongStudentRecord("에러가 발생했습니다. 크롤링을 다시 시작해주세요")
         
```
