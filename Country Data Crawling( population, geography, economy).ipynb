```python
from selenium import webdriver
from bs4 import BeautifulSoup
import urllib.request
import re   
import time
import numpy as np
import pandas
import csv


word = '</strong>'
num = 2


class population_crawling: 
    
    def median_age():
        #open file 'median_age.csv' and record crawled data 
        with open('median_age.csv', 'w', encoding='UTF-8', newline='') as f:
            writer = csv.writer(f)
            writer.writerow( ['country name', 'total','male', 'female'] )

            pf= pandas.read_csv('CountryNameCode.csv', engine='python')

            for i in range(30):
                #open'CountryNameCode.csv' and read 0 column data from row 1~30
                
                mtitle=pf[pf.columns[0]].iloc[i]


                url='https://www.indexmundi.com/'+str(mtitle)+'/demographics_profile.html'

                print(url)
                
                #regular expression 
                total = re.compile(r'.<strong>total: </strong>(.*)<br/><strong>male.*') 
                male=re.compile(r'.<strong>male: </strong>(.*)<br/><strong>female.*')
                female=re.compile(r'.<strong>female: </strong>(.*)<br/>.*')

                driver=webdriver.Chrome("C:/Users/ATIV/Downloads/chromedriver_win32/chromedriver")
                
                #open url 
                driver.get(url)
                soup = BeautifulSoup(driver.page_source, "html.parser")


                link = soup.find_all("div", {"class":"col-md-8 c"})
                print("===================================")

                for posts in link:
                    time.sleep(0.33)
                    z=str(posts)
                    zz=str(posts).split(' ')
                    #print(str(zz))
                    str_index=0
                    count = 0
                    for aaa in zz:
                        #index = 0
                        str_index+=len(aaa)
                        if word in aaa:  
                            count+=1
                            if(count == num):

                                index=aaa.find(word[len(word)-1])
                                print(z[str_index-len(aaa)+len(word):])                #search word가 붙어있을 경우     ex) </strong>12.84%
                                #print(z[str_index:])                                    #search word가 띄어져 있을 경우  ex) </strong> 12.84%
                                break
                        str_index+=1

                    a=total.search(z).group(1).strip()
                    b=male.search(z).group(1).strip()
                    c=female.search(z).group(1).strip()
               
                #save country data in each cell   
                Field1=mtitle   #country name 
                Field2=a.split()[0] 
                Field3=b.split()[0]
                Field4=c.split()[0]
               
             
                writer.writerow([Field1,Field2,Field3,Field4])

    def age_structure():
        
        #open file 'age_structure.csv' and record crawled data 
        with open('age_structure.csv', 'w', encoding='UTF-8', newline='') as f:
            writer = csv.writer(f)
            writer.writerow( ['country name','0-14', '15-24', '25-54', '54-64', '65+'] )

            pf= pandas.read_csv('CountryNameCode.csv', engine='python')

            for i in range(30):

                mtitle=pf[pf.columns[0]].iloc[i]
                #open'CountryNameCode.csv' and read 0 column data from row 1~30


                url='https://www.indexmundi.com/'+str(mtitle)+'/age_structure.html'
                #make 30 url to crawl data

                print(url)
                
                #regular expression for crwaling

                Age0_14=re.compile(r'.0-14 years: </strong>(.*)<br/><strong>15.*')
                Age15_24=re.compile(r'.15-24 years: </strong>(.*)<br/><strong>25.*')
                Age25_54=re.compile(r'.25-54 years: </strong>(.*)<br/><strong>55.*')
                Age54_64=re.compile(r'.55-64 years: </strong>(.*)<br/><strong>65.*')
                Age65=re.compile(r'.65 years and over: </strong>(.*)<br/>.*')

                driver=webdriver.Chrome("C:/Users/ATIV/Downloads/chromedriver_win32/chromedriver")
                driver.get(url)
                soup = BeautifulSoup(driver.page_source, "html.parser")

                link = soup.find_all("div", {"class":"c"})
                print("===================================")

                for posts in link:
                    time.sleep(0.33)
                    z=str(posts)
                    zz=str(posts).split(' ')
                    #print(str(zz))
                    str_index=0
                    count = 0
                    for aaa in zz:
                        #index = 0
                        str_index+=len(aaa)
                        if word in aaa:  
                            count+=1
                            if(count == num):

                                index=aaa.find(word[len(word)-1])
                                print(z[str_index-len(aaa)+len(word):])                #search word가 붙어있을 경우     ex) </strong>12.84%
                                print(z[str_index:])                                    #search word가 띄어져 있을 경우  ex) </strong> 12.84%
                                break
                        str_index+=1

                    a=Age0_14.search(z).group(1).strip()
                    b=Age15_24.search(z).group(1).strip()
                    c=Age25_54.search(z).group(1).strip()
                    d=Age54_64.search(z).group(1).strip()
                    e=Age65.search(z).group(1).strip()
                    
                #save data in each cell after delete'%'sign 
                Field1=mtitle
                Field3=a.split()[0][:len(a.split()[0])-1]
                Field4=b.split()[0][:len(b.split()[0])-1]
                Field5=c.split()[0][:len(c.split()[0])-1]
                Field6=d.split()[0][:len(d.split()[0])-1]
                Field7=e.split()[0][:len(e.split()[0])-1]
                
             
                writer.writerow([Field1,Field3,Field4,Field5, Field6, Field7])

class economy_crawling:
    
    def Unemployment():
        #open file 'unemployment.csv' and record crawled data 
        with open('Unemployment.csv', 'w', encoding='UTF-8', newline='') as f:
            writer = csv.writer(f)
            writer.writerow( ['country name','2006','2007','2008','2009', '2010','2011', '2012','2013','2014'] )

            pf= pandas.read_csv('CountryNameCode.csv', engine='python')
            
            #open 'CountryNameCode.csv' and read column 1 from row 14-29 
            #b/c several countries data don't have time series unemployment rate (2006~2014)

            for i in range(14,30):

                mtitle=pf[pf.columns[0]].iloc[i]


                url='https://www.indexmundi.com/'+str(mtitle)+'/unemployment_rate.html'
                #make url to crawl data

                print(url)
                
                #regular expression for data crawling by time series
                unemp_2006=re.compile(r'.2006</td><td>(.*)</td><td>.*') 
                unemp_2007=re.compile(r'.2007</td><td>(.*)</td><td>.*') 
                unemp_2008=re.compile(r'.2008</td><td>(.*)</td><td>.*') 
                unemp_2009=re.compile(r'.2009</td><td>(.*)</td><td>.*') 
                unemp_2010=re.compile(r'.2010</td><td>(.*)</td><td>.*') 
                unemp_2011=re.compile(r'.2011</td><td>(.*)</td><td>.*')
                unemp_2012=re.compile(r'.2012</td><td>(.*)</td><td>.*')
                unemp_2013=re.compile(r'.2013</td><td>(.*)</td><td>.*')
                unemp_2014=re.compile(r'.2014</td><td>(.*)</td><td>.*')
              
                driver=webdriver.Chrome("C:/Users/ATIV/Downloads/chromedriver_win32/chromedriver")
                driver.get(url)
                soup = BeautifulSoup(driver.page_source, "html.parser")


                link = soup.find_all("table", {"class":"tblrank"})
       
                print("===================================")

                for posts in link:
                    time.sleep(0.33)
                    z=str(posts)
                    zz=str(posts).split(' ')
                    #print(str(zz))
                    str_index=0
                    count = 0
                    for aaa in zz:
                        #index = 0
                        str_index+=len(aaa)
                        if word in aaa:  
                            count+=1
                            if(count == num):

                                index=aaa.find(word[len(word)-1])
                                print(z[str_index-len(aaa)+len(word):])                #search word가 붙어있을 경우     ex) </strong>12.84%
                                print(z[str_index:])                                    #search word가 띄어져 있을 경우  ex) </strong> 12.84%
                                break
                        str_index+=1

                    
                    a=unemp_2006.search(z).group(1).strip()
                    b=unemp_2007.search(z).group(1).strip()
                    c=unemp_2008.search(z).group(1).strip()
                    d=unemp_2009.search(z).group(1).strip()
                    e=unemp_2010.search(z).group(1).strip()
                    f=unemp_2011.search(z).group(1).strip()
                    g=unemp_2012.search(z).group(1).strip()
                    h=unemp_2013.search(z).group(1).strip()
                    i=unemp_2014.search(z).group(1).strip()
                
                #save selected data in each cell
                Field1=mtitle
                Field2=a.split()[0][:3]
                Field3=b.split()[0][:3]
                Field4=c.split()[0][:3]
                Field5=d.split()[0][:3]
                Field6=e.split()[0][:3]
                Field7=f.split()[0][:3]
                Field8=g.split()[0][:3]
                Field9=h.split()[0][:3]
                Field10=i.split()[0][:3]
                
             
                writer.writerow([Field1,Field2,Field3,Field4,Field5,Field6,Field7,Field8,Field9,Field10])
                
    def GDP_composition():
        #open file 'GDP_composition.csv' and record crawled data / encoding_ utf8
        with open('GDP_composition.csv', 'w', encoding='UTF-8', newline='') as f:
            writer = csv.writer(f)
            #make title of rows _ as 6 components of GDP
            writer.writerow( ['country name', 'household consumption','government consumption', 'investment in fixed capital',\
            'investment in inventories','exports of S&G','imports of S&G'] )

            pf= pandas.read_csv('CountryNameCode.csv', engine='python')

            for i in range(30):
                
                #read 30 countries name from 'CountryNameCode.csv' and make 30 urls

                mtitle=pf[pf.columns[0]].iloc[i]


                url='https://www.indexmundi.com/'+str(mtitle)+'/economy_profile.html'

                print(url)
                
                #regular expression for crwaling 
                HS_consumption = re.compile(r'.household consumption: </strong>(.*)<br/><strong>government.*') 
                Gov_consumption=re.compile(r'.government consumption: </strong>(.*)<br/><strong>investment in fixed capital.*')
                inv_fixedcapital=re.compile(r'.investment in fixed capital: </strong>(.*)<br/><strong>investment in inventories.*')
                inv_inventories=re.compile(r'.inventories: </strong>(.*)<br/><strong>exports of goods and services.*')
                exports = re.compile(r'.exports of goods and services: </strong>(.*)<br/><strong>imports.*')
                imports = re.compile(r'.imports of goods and services: </strong>(.*)<br/>.*')

                driver=webdriver.Chrome("C:/Users/ATIV/Downloads/chromedriver_win32/chromedriver")
                driver.get(url)
                soup = BeautifulSoup(driver.page_source, "html.parser")


                link = soup.find_all("div", {"class":"col-md-8 c"})
    
                print("===================================")

                for posts in link:
                    time.sleep(0.33)
                    z=str(posts)
                    zz=str(posts).split(' ')
                    #print(str(zz))
                    str_index=0
                    count = 0
                    for aaa in zz:
                        #index = 0
                        str_index+=len(aaa)
                        if word in aaa:  
                            count+=1
                            if(count == num):

                                index=aaa.find(word[len(word)-1])
                                print(z[str_index-len(aaa)+len(word):])                #search word가 붙어있을 경우     ex) </strong>12.84%
                                print(z[str_index:])                                    #search word가 띄어져 있을 경우  ex) </strong> 12.84%
                                break
                        str_index+=1

                    a=HS_consumption.search(z).group(1).strip()
                    b=Gov_consumption.search(z).group(1).strip()
                    c=inv_fixedcapital.search(z).group(1).strip()
                    d=inv_inventories.search(z).group(1).strip()
                    e=exports.search(z).group(1).strip()
                    f=imports.search(z).group(1).strip()
                    
                Field1=mtitle
                
                #save each data without '%' sign in each cell
                Field2=a.split()[0][:len(a.split()[0])-1]
                Field3=b.split()[0][:len(b.split()[0])-1]
                Field4=c.split()[0][:len(c.split()[0])-1]
                Field5=d.split()[0][:len(d.split()[0])-1]
                Field6=e.split()[0][:len(e.split()[0])-1]
                Field7=f.split()[0][:len(f.split()[0])-1]
               
             
                writer.writerow([Field1,Field2,Field3,Field4,Field5,Field6,Field7])
                
    


    

```

```python
class geography_crawling: 
    
    def area_comp():
        #open file 'area_comp.csv' and record crawled data / encoding_ utf8
        with open('area_comp.csv', 'w', encoding='UTF-8', newline='') as f:
            writer = csv.writer(f)
            
            #set title of each column 
            writer.writerow( ['country name', 'total','land', 'water'] )

            pf= pandas.read_csv('CountryNameCode.csv', engine='python')

            for i in range(30):

                mtitle=pf[pf.columns[0]].iloc[i]

                #make url of each country 
                url='https://www.indexmundi.com/'+str(mtitle)+'/geography_profile.html'

                print(url)
                
                #regular expression for crawling 
                total = re.compile(r'.total: </strong>(.*)<br/><strong>land.*') 
                land=re.compile(r'.land: </strong>(.*)<br/><strong>water.*')
                water=re.compile(r'.water: </strong>(.*)<br/>.*')

                driver=webdriver.Chrome("C:/Users/ATIV/Downloads/chromedriver_win32/chromedriver")
                
                #get url 
                driver.get(url)
                soup = BeautifulSoup(driver.page_source, "html.parser")


                link = soup.find_all("div", {"class":"col-md-8 c"})
                print("===================================")

                for posts in link:
                    time.sleep(0.33)
                    z=str(posts)
                    zz=str(posts).split(' ')
                    #print(str(zz))
                    str_index=0
                    count = 0
                    for aaa in zz:
                        #index = 0
                        str_index+=len(aaa)
                        if word in aaa:  
                            count+=1
                            if(count == num):

                                index=aaa.find(word[len(word)-1])
                                print(z[str_index-len(aaa)+len(word):])                #search word가 붙어있을 경우     ex) </strong>12.84%
                                print(z[str_index:])                                    #search word가 띄어져 있을 경우  ex) </strong> 12.84%
                                break
                        str_index+=1

                    a=total.search(z).group(1).strip()
                    b=land.search(z).group(1).strip()
                    c=water.search(z).group(1).strip()
               
            # save data in each cell 
                Field1=mtitle
                Field2=a.split()[0]
                Field3=b.split()[0]
                Field4=c.split()[0]
               
             
                writer.writerow([Field1,Field2,Field3,Field4])

    def land_use():
        #open file 'land_use.csv' and record crawled data / encoding_ utf8
        with open('land_use.csv', 'w', encoding='UTF-8', newline='') as f:
            writer = csv.writer(f)
            
            #set title of each column 
            writer.writerow( ['country name','agriculture_land','forest','other','total'] )

            pf= pandas.read_csv('CountryNameCode.csv', engine='python')

            for i in range(30):

                mtitle=pf[pf.columns[0]].iloc[i]


                url='https://www.indexmundi.com/'+str(mtitle)+'/geography_profile.html'

                print(url)
                
                
                #regular expression for crawling 
                agri=re.compile(r'.agricultural land: </strong>(.*)<br/>arable.*')
                forest=re.compile(r'.forest: </strong>(.*)<br/><strong>other.*')
                other=re.compile(r'.other: </strong>(.*)<br/></td></tr>.*')
                
                driver=webdriver.Chrome("C:/Users/ATIV/Downloads/chromedriver_win32/chromedriver")
                
                #get url using chrome driver
                driver.get(url)
                soup = BeautifulSoup(driver.page_source, "html.parser")

                link = soup.find_all("div", {"class":"col-md-8 c"})
                print("===================================")

                for posts in link:
                    time.sleep(0.33)
                    z=str(posts)
                    zz=str(posts).split(' ')
                    #print(str(zz))
                    str_index=0
                    count = 0
                    for aaa in zz:
                        #index = 0
                        str_index+=len(aaa)
                        if word in aaa:  
                            count+=1
                            if(count == num):

                                index=aaa.find(word[len(word)-1])
                                print(z[str_index-len(aaa)+len(word):])                #search word가 붙어있을 경우     ex) </strong>12.84%
                                print(z[str_index:])                                    #search word가 띄어져 있을 경우  ex) </strong> 12.84%
                                break
                        str_index+=1

                    a=agri.search(z).group(1).strip()
                    b=forest.search(z).group(1).strip()
                    c=other.search(z).group(1).strip()
                    
                #save data in each cell 
                Field1=mtitle
                Field2=a.split()[0][:len(a.split()[0])-1]
                Field3=b.split()[0][:len(b.split()[0])-1]
                Field4=c.split()[0][:len(c.split()[0])-1]
                Field5 = 100
                
             
                writer.writerow([Field1,Field2,Field3,Field4,Field5])
```

```python
def CheckTime(func):
    #checktime function to check total time for crawling 
    import time

    def newFunc(*args, **kwargs):
        start = time.time()
        func(*args, **kwargs)
        end = time.time()
        
        print("\n실행시간 : ",end - start)

    return newFunc
```

```python
class WrongDataRecord(Exception):
    #exception handing _error message 
    def __init__(self, msg):
        self._message = msg
```

```python
#exception
@CheckTime
def get_population():
    population_crawling.age_structure()
    population_crawling.median_age()

@CheckTime
def get_Unemp():  
    try: 
        economy_crawling.Unemployment()
    except AttributeError:
        pass
        
    except IndexError:
        raise WrongDataRecord("CountryNameCode.csv 파일의 나라 개수를 다시 확인해주세요")
    
    except:
        pass 


@CheckTime
def get_GDP():
    try:
        economy_crawling.GDP_composition()
    except:
        raise WrongDataRecord("에러가 발생했습니다. 크롤링을 다시 시작해주세요")

@CheckTime
def get_GEO():
    try:
        geography_crawling.area_comp()
        geography_crawling.land_use()
    except:
        raise WrongDataRecord("에러가 발생했습니다. 크롤링을 다시 시작해주세요")
        
        
```
